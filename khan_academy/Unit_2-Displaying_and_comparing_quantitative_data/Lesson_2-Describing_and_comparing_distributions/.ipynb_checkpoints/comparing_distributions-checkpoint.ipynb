## 1. Semifinal Round Observations

  - 3 swimmers finished at **53.5 seconds**.

  - Fastest swimmer at **52.7 seconds**.

## 2. Final Round Observations

  - One swimmer much faster: **52.2 seconds**.  
  - One swimmer slower: **53.8 seconds** (worse than all semifinal times).

## 3. Statements Evaluated
- âœ… **Swimmers had faster times on average in the finals.**  
  - Mean time in finals â‰ˆ 53.1â€“53.2 seconds.  
  - Mean time in semifinals â‰ˆ 53.3+ seconds.  
  - â†’ Average improved.  

- âŒ **One swimmer was disqualified.**  
  - False: 8 swimmers in both rounds.  

- âœ… **Times in the finals varied more than in the semifinals.**  
  - Semifinals: tightly clustered around 53.3â€“53.5.

- âŒ **Individually, all swimmers swam faster in the finals.**

---

## âœ… Key Takeaways
- **Average performance improved** in the finals.  
- **Variation increased** in the finals (greater spread of times).  
- **Not all swimmers improved individually** â€” one swimmer slowed down.  
- **No disqualifications** occurred.


```python
# Program to compare semifinal and final times in Olympic swimming
import statistics
import matplotlib.pyplot as plt

# Step 1: Define datasets (times in seconds)
semifinal_times = [53.5, 53.5, 53.5, 53.7, 52.7, 53.3, 53.4, 53.6]  # example semifinal times
final_times = [52.2, 53.8, 53.1, 53.2, 53.3, 53.4, 53.5, 53.6]      # example final times

# Step 2: Calculate descriptive statistics
def describe_data(data, label):
    mean_val = statistics.mean(data)
    variance_val = statistics.pvariance(data)
    min_val = min(data)
    max_val = max(data)
    print(f"{label} -> Mean: {mean_val:.2f}, Variance: {variance_val:.4f}, Range: {min_val}-{max_val}")

describe_data(semifinal_times, "Semifinals")
describe_data(final_times, "Finals")

# Step 3: Visualize dot plots
plt.figure(figsize=(8,5))

# Semifinal dot plot
plt.scatter(semifinal_times, [1]*len(semifinal_times), color="blue", label="Semifinals")

# Final dot plot
plt.scatter(final_times, [0]*len(final_times), color="red", label="Finals")

plt.yticks([0,1], ["Finals", "Semifinals"])
plt.xlabel("Time (seconds)")
plt.title("Olympic 100m Backstroke Times (2012)")
plt.legend()
plt.grid(axis='x', linestyle='--', alpha=0.6)
```

# ðŸ”Ž What this program does:

1. Stores semifinal and final times for 8 swimmers.
2. Computes statistics:

    - **Mean â†’** average performance.
    - **Variance â†’** spread of times.
    - **Range â†’** fastest vs slowest.

3. Plots dot plots:

    - **Blue dots =** semifinal times.
    - **Red dots =** final times.
    - Easy to compare distributions visually.


# ðŸ’¡ Running this program will show:

* **Mean finals < mean semifinals â†’** swimmers were faster on average in finals.
* **Variance finals > variance semifinals â†’** more spread in final times.
* Dot plot visualization makes it clear which swimmers improved and which slowed down.
