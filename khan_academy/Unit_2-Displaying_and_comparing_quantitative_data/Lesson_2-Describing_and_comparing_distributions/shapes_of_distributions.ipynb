# üìä Shapes of Distributions

## 1. Right-Tailed (Right-Skewed) Distribution
- **Example:** Matt‚Äôs Caf√© guest ages  
- Characteristics:
  - No guests under 20
  - Many guests between 30‚Äì40
  - Fewer guests as age increases beyond 50, with small counts up to 90
- Shape: Bulk of data on the left, **tail extends to the right**
- Interpretation: **Right-tailed / Right-skewed**

---

## 2. Symmetrical Distribution
- **Example:** Logan Assisted Living guest ages  
- Characteristics:
  - Many guests between 60‚Äì70
  - Reasonable counts between 50‚Äì60 and 70‚Äì80
- Shape: Roughly balanced around age 65
- Interpretation: **Approximately symmetrical**

---

## 3. Left-Tailed (Left-Skewed) Distribution
- **Example:** Box-and-whisker plot dataset  
- Characteristics:
  - Minimum = 11, Maximum = 25, Median = 21
  - Box shows middle 50% of values
- Shape: Longer whisker/tail on the **left side**
- Interpretation: **Skewed to the left**

---

## 4. Key Vocabulary
- **Right-tailed / Right-skewed**: Tail extends to the right; mean > median  
- **Left-tailed / Left-skewed**: Tail extends to the left; mean < median  
- **Symmetrical**: Both sides of the distribution mirror each other  
- **Skewness**: A measure of asymmetry in the distribution

---

## ‚úÖ Takeaway
- Distributions can be described by their **shape**: right-tailed, left-tailed, or symmetrical  
- Tails indicate where fewer values stretch out  
- Symmetry means the data is balanced around a central point  
- Skewness connects tails to the relationship between mean, median, and mode


```python
# Program to analyze distribution shape: mean, median, mode, skewness
import statistics
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import skew

# Example dataset (ages of guests at a cafe)
data = [22, 25, 28, 30, 31, 32, 33, 35, 40, 45, 50, 55, 60, 65, 70, 75, 80]

# Step 1: Calculate measures of central tendency
mean_value = statistics.mean(data)
median_value = statistics.median(data)
try:
    mode_value = statistics.mode(data)
except statistics.StatisticsError:
    mode_value = "No unique mode"

# Step 2: Calculate skewness
skewness_value = skew(data)

# Step 3: Print results
print("Dataset:", data)
print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)
print("Skewness:", skewness_value)

if skewness_value > 0:
    print("‚û°Ô∏è Distribution is Right-Skewed (Right-Tailed)")
elif skewness_value < 0:
    print("‚û°Ô∏è Distribution is Left-Skewed (Left-Tailed)")
else:
    print("‚û°Ô∏è Distribution is Approximately Symmetrical")

# Step 4: Plot histogram
plt.figure(figsize=(8, 5))
sns.histplot(data, bins=8, kde=True, color="skyblue", edgecolor="black")

plt.axvline(mean_value, color='red', linestyle='--', label=f"Mean = {mean_value:.2f}")
plt.axvline(median_value, color='green', linestyle='--', label=f"Median = {median_value:.2f}")

plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Distribution Shape Analysis")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

# üîé What this program does

1. Computes **mean, median, and mode** for the dataset.
2. Calculates **skewness** using **scipy.stats.skew**.

    - Positive skew ‚Üí right-tailed
    - Negative skew ‚Üí left-tailed
    - Near zero ‚Üí symmetrical

3. Prints interpretation of the distribution shape.
4. Plots a **histogram** with mean and median lines for visual comparison.
