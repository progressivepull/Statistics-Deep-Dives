# üìä Data Representations: Dot Plot, Histogram, and Box Plot

## 1. Pixar Film Running Times
- **Dot Plot**  
  - Shows each film individually.  
  - Can list all running times, order them, and find the **median**.  
  - Useful for calculating the median manually.  

- **Histogram**  
  - Groups films into intervals (e.g., 80‚Äì85 minutes).  
  - Exact values are hidden ‚Üí cannot determine the median precisely.  
  - **Not useful** for finding the median.  

- **Box Plot**  
  - Explicitly shows the **median** with the line inside the box.  
  - Median here = **99 minutes**.  
  - **Most efficient** for finding the median.

---

## 2. Used Car Lot Mileage
- **Question 1: How many vehicles had driven more than 200,000 km?**  
  - **Histogram**: Shows frequency in ranges.  
    - 3 vehicles between 200k‚Äì250k.  
    - 2 vehicles between 250k‚Äì300k.  
    - Total = **5 vehicles**.  
    - ‚úÖ Histogram is useful.  
  - **Box Plot**: Only shows quartiles, not counts.  
    - ‚ùå Not useful for this question.  

- **Question 2: What was the median mileage?**  
  - **Histogram**: Groups values, hides exact middle ‚Üí ‚ùå Not useful.  
  - **Box Plot**: Median line clearly shown at **140,000 km**.  
    - ‚úÖ Box plot is useful.

---

## ‚úÖ Key Takeaways
- **Dot plots**: Best for listing all values and manually finding measures like the median.  
- **Histograms**: Best for understanding **frequency counts** and ranges (e.g., how many items fall above/below a threshold).  
- **Box plots**: Best for quickly identifying the **median, quartiles, and spread** of data.


```python
# Program to visualize data using dot plot, histogram, and box plot
import matplotlib.pyplot as plt
import seaborn as sns

# Example dataset: running times of Pixar's first 14 films (in minutes)
data = [81, 92, 93, 95, 96, 96, 98, 100, 102, 104, 105, 107, 108, 111]

# Step 1: Dot Plot
plt.figure(figsize=(12,4))
sns.stripplot(data, size=8, color="blue", jitter=False)
plt.title("Dot Plot of Film Running Times")
plt.xlabel("Running Time (minutes)")
plt.show()

# Step 2: Histogram
plt.figure(figsize=(12,4))
plt.hist(data, bins=6, color="skyblue", edgecolor="black")
plt.title("Histogram of Film Running Times")
plt.xlabel("Running Time (minutes)")
plt.ylabel("Frequency")
plt.show()

# Step 3: Box Plot
plt.figure(figsize=(6,4))
sns.boxplot(data, color="lightgreen")
plt.title("Box Plot of Film Running Times")
plt.xlabel("Running Time (minutes)")
plt.show()
```

# üîé What this program does:

* **Dot Plot ‚Üí** Shows each film individually, useful for listing values and finding the **median manually**.
* **Histogram ‚Üí** Groups films into bins, useful for seeing **frequency counts** and ranges.
* **Box Plot ‚Üí** Summarizes distribution, explicitly shows **median, quartiles, and spread**.
