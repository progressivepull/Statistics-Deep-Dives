  - Vertical axis â†’ stock price.  
  - Data points connected by lines to highlight **trends over time**.  
- Line graphs are especially useful for showing **changes and overall direction**.

## 2. Month-to-Month Movements
- July â†’ price â‰ˆ $10.  
- August â†’ price rose to â‰ˆ $11.  
- September â†’ price dipped.  
- Octoberâ€“November â†’ price rose again.

- March â†’ sharp rise to â‰ˆ $17.

## 3. Overall Trend
- Despite short-term ups and downs, the **general direction is upward**.  
- Start: â‰ˆ $10 in July.  
- End: â‰ˆ $16â€“17 in June.  
- âœ… Conclusion: The stock price is **rising over the course of the year**.

## âœ… Key Takeaway
- Line graphs show **trends over time**.  
- Even with temporary declines, the **overall trajectory matters most**.  
- Here, the stockâ€™s **long-term trend is upward growth**.


```python
# Program to analyze stock performance over 12 months
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Define months and stock prices (example data)
months = ["Jul", "Aug", "Sep", "Oct", "Nov", "Dec", 
          "Jan", "Feb", "Mar", "Apr", "May", "Jun"]

# Example stock prices (in dollars)
prices = [10.2, 11.0, 10.5, 11.3, 12.0, 11.7, 
          12.5, 13.2, 16.8, 15.5, 16.0, 16.7]

# Step 2: Calculate overall trend
start_price = prices[0]
end_price = prices[-1]
trend = "rising" if end_price > start_price else "falling" if end_price < start_price else "staying the same"

print("Stock Performance Analysis")
print("--------------------------")
print(f"Start Price (July): ${start_price}")
print(f"End Price (June): ${end_price}")
print(f"Overall Trend: The stock is {trend} over the year.")

# Step 3: Plot line graph
plt.figure(figsize=(10,5))
plt.plot(months, prices, marker="o", linestyle="-", color="blue", linewidth=2)

# Highlight start and end points
plt.scatter(months[0], start_price, color="green", s=100, label=f"Start: ${start_price}")
plt.scatter(months[-1], end_price, color="red", s=100, label=f"End: ${end_price}")

plt.title("Stock Performance Over 12 Months")
plt.xlabel("Month")
plt.ylabel("Stock Price ($)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()
plt.show()
```

# ðŸ”Ž What this program does:

* **Stores monthly stock prices** in a list.
* **Compares start vs end price** to determine if the stock is rising, falling, or staying the same.
* **Prints analysis** of the trend.
* **Plots a line graph** with markers for each month, highlighting start and end prices.
