# Misleading line graphs

# ü•§ Thrill Soda vs Yummy Cola Graphs

## 1. Data Collected
- **2006:** Yummy Cola 80%, Thrill Soda 12%, No preference 8%.  
- **2007‚Äì2011:** Yummy Cola gradually declined (76% ‚Üí 68%).  
- **2007‚Äì2011:** Thrill Soda gradually increased (12% ‚Üí 25%).  
- ‚úÖ Raw data is accurate and faithfully represented.

---

## 2. Graphs Presented
- **Yummy Cola graph:** Shows a downward trend (80% ‚Üí 68%).  
- **Thrill Soda graph:** Shows an upward trend (12% ‚Üí 25%).  
- Both graphs individually are correct.  

---

## 3. Misleading Aspects
- **Different scales used:**  
  - Yummy Cola graph starts at **50%**, not 0%.  
  - Thrill Soda graph starts at **0%** with a compressed scale (0‚Äì30%).  

- **Visual distortion:**  
  - Thrill Soda‚Äôs upward trend appears visually ‚Äúhigher‚Äù than Yummy Cola‚Äôs downward trend.  
  - Human brain tempted to compare across graphs, even though scales differ.  

- **Impact:**  
  - Makes Thrill Soda look like it is catching up or surpassing Yummy Cola.  
  - Reality: Yummy Cola still has far more preference every year.

---

## 4. Honest Representation
- Plotting both brands on the **same graph with the same scale** would show:  
  - Yummy Cola consistently dominant (68‚Äì80%).  
  - Thrill Soda much lower (12‚Äì25%).  
- The upward trend for Thrill Soda is real, but **not comparable** to Yummy Cola‚Äôs majority.

---

## ‚úÖ Key Takeaways
- The graphs are **accurate but misleading** due to scale manipulation.  
- Line graphs can exaggerate trends if scales are inconsistent.  
- Always check **axes and scales** before interpreting comparisons.  
- Reality: Yummy Cola remains far more popular, despite Thrill Soda‚Äôs growth.


```python
# Program to plot Thrill Soda vs Yummy Cola preferences on the same scale
import matplotlib.pyplot as plt

# Step 1: Define years and survey data
years = [2006, 2007, 2008, 2009, 2010, 2011]
yummy = [80, 76, 77, 73, 73, 68]   # % preferring Yummy Cola
thrill = [12, 19, 19, 20, 21, 25]  # % preferring Thrill Soda
no_pref = [8, 5, 4, 7, 6, 7]       # % with no preference

# Step 2: Plot both lines on the same graph
plt.figure(figsize=(10,6))
plt.plot(years, yummy, marker="o", color="red", label="Yummy Cola")
plt.plot(years, thrill, marker="o", color="blue", label="Thrill Soda")
plt.plot(years, no_pref, marker="o", color="gray", linestyle="--", label="No Preference")

# Step 3: Formatting
plt.title("Consumer Soda Preferences (2006‚Äì2011)")
plt.xlabel("Year")
plt.ylabel("Percentage of Respondents")
plt.ylim(0,100)  # same scale for all lines
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()
plt.show()

# Step 4: Print summary stats
print("Survey Results (2006‚Äì2011):")
for i in range(len(years)):
    print(f"{years[i]}: Yummy={yummy[i]}%, Thrill={thrill[i]}%, No Pref={no_pref[i]}%")

# Step 5: Trend analysis
if thrill[-1] > thrill[0]:
    print("\nThrill Soda preference increased over time.")
if yummy[-1] < yummy[0]:
    print("Yummy Cola preference decreased over time.")
print("But Yummy Cola remained far more popular every year.")
```

# üîé What this program does:

1. Stores **survey data** for Yummy Cola, Thrill Soda, and No Preference.
2. Plots all three lines on the **same scale (0‚Äì100%)** to avoid misleading comparisons.
3. Prints year-by-year percentages.
4. Summarizes the **trend**:

    - Thrill Soda increased (12% ‚Üí 25%).
    - Yummy Cola decreased (80% ‚Üí 68%).
    - But Yummy Cola remained dominant every year.
