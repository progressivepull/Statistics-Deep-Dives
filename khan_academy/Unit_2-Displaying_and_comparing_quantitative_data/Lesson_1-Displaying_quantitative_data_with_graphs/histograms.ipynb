## 1. What is a Histogram?
- A **histogram** is a way of visualizing data.
- It looks like a bar graph, but instead of plotting raw values, it shows the **frequency** of values (how often each number or range appears).
- Itâ€™s one of the most common tools in statistics for representing data distributions.

## 2. Example Dataset
Numbers collected:  
`0, 1, 4, 2, 1, 0, 2, 2, 1, 0, 1, 2, 1, 0, 0, 2, 2, 3, 1, 1, 3, 6`

## 3. Frequency Counts
| Number | Frequency |
|--------|-----------|
| 0      | 4 |
| 1      | 7 |
| 2      | 5 |
| 3      | 2 |
| 4      | 1 |
| 5      | 0 |
| 6      | 1 |

- **0** appears 4 times  
- **1** appears 7 times  
- **2** appears 5 times  
- **3** appears 2 times  
- **4** appears 1 time  
- **5** appears 0 times  
- **6** appears 1 time  

## 4. Histogram Construction
- **X-axis** â†’ The numbers (0 through 6).  
- **Y-axis** â†’ Frequency (how many times each number appears).  
- Each barâ€™s height corresponds to the frequency of that number.  
  - Example: The bar for **1** reaches up to 7, since 1 appears 7 times.

## 5. Key Insight
- A histogram is simply a **visual representation of frequency counts**.  
- It makes patterns in the data easy to see:
  - Which values occur most often (here, **1** is the most frequent).
  - Which values are rare (like **4** and **6**).
  - Which values donâ€™t appear at all (**5**).


```python
# Histogram Example: Frequency of Numbers
import matplotlib.pyplot as plt
from collections import Counter

# Step 1: Dataset (from your example)
data = [0, 1, 4, 2, 1, 0, 2, 2, 1, 0, 1, 2, 1, 0, 0, 
        2, 2, 3, 1, 1, 3, 6]

# Step 2: Frequency table using Counter
freq_table = Counter(data)

print("Frequency Table:")
print("Number | Frequency")
for num in range(min(data), max(data) + 1):
    print(f"{num:>6} | {freq_table.get(num, 0)}")

# Step 3: Plot histogram
plt.figure(figsize=(8, 5))
plt.bar(freq_table.keys(), freq_table.values(), color='skyblue', edgecolor='black')

# Step 4: Labeling
plt.xlabel("Number")
plt.ylabel("Frequency")
plt.title("Histogram of Data Frequencies")
plt.xticks(range(min(data), max(data) + 1))
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Step 5: Show plot
plt.show()
```
