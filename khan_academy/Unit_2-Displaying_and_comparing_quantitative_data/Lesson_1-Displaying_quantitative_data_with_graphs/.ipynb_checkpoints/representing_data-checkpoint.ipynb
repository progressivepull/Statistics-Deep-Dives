## Data Definition
Data is recorded information. It can be:
- **Quantitative**: numeric scores, percentages.
- **Qualitative**: opinions, ratings, categories.

---

## Example Dataset
Student test scores:

| Name  | Score |
|-------|-------|
| Amy   | 90    |
| Bill  | 95    |
| Cam   | 100   |
| Efra  | 100   |
| Farah | 80    |

---

## Representations of Data
1. **Table format**  
   - Names in one column, scores in another.  
   - Common in databases.

2. **Bar graph / histogram**  
   - Names on the x-axis, scores on the y-axis.  
   - Visual comparison of scores.

3. **List of scores**  
   - Just the numbers: `90, 95, 100, 100, 80`.  
   - Less information since names are missing.

4. **Frequency plot**  
   - Scores on the x-axis, frequency (count) on the y-axis.  
   - Shows how often each score occurs.

---

## Questions You Can Ask With Data
- **Range**:  
  
$$
\text{Max} - \text{Min} = 100 - 80 = 20
$$

- **Count below 100**:  
  3 students.

- **Most frequent score**:  
  100 (appears twice).

---

## Key Takeaway
All these methods represent the **same underlying data** but highlight different aspects:
- Tables emphasize detail.  
- Graphs emphasize visualization.  
- Frequency plots emphasize distribution.  

Each format allows us to ask and answer meaningful questions.


```python
# Representing student scores in different formats

import matplotlib.pyplot as plt
from collections import Counter

# Example dataset
scores = {
    "Amy": 90,
    "Bill": 95,
    "Cam": 100,
    "Efra": 100,
    "Farah": 80
}

# 1. Table format (dictionary printout)
print("Table of Scores:")
for name, score in scores.items():
    print(f"{name:5} | {score}")

# 2. List of scores
score_list = list(scores.values())
print("\nList of Scores:", score_list)

# 3. Bar graph
plt.figure(figsize=(6,4))
plt.bar(scores.keys(), scores.values(), color="skyblue")
plt.title("Student Test Scores")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.ylim(0, 110)
plt.show()

# 4. Frequency plot
freq = Counter(score_list)
plt.figure(figsize=(6,4))
plt.bar(freq.keys(), freq.values(), color="lightgreen")
plt.title("Frequency of Scores")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.show()

# 5. Interesting questions
range_scores = max(score_list) - min(score_list)
below_100 = sum(1 for s in score_list if s < 100)
most_frequent = max(freq, key=freq.get)

print("\nData Analysis:")
print(f"Range of scores: {range_scores}")
print(f"Number of students scoring below 100: {below_100}")
print(f"Most frequent score: {most_frequent}")
```
