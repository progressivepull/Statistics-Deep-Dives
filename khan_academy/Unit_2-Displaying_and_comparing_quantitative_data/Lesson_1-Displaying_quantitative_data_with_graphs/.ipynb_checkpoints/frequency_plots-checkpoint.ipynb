# ðŸ“Š Summary: Representing Class Ages with Frequency Tables & Dot Plots

## 1. Data Representation
- **List of ages**: Raw data showing each studentâ€™s age.
- **Frequency table**: Organizes ages into two columns:
  - Age
  - Number of students (frequency)
- **Dot plot**: Visual representation where each dot corresponds to one student at a given age.

## 2. Constructed Frequency Table
| Age | Number of Students (Frequency) |
|-----|-------------------------------|
| 5   | 2 |
| 6   | 1 |
| 7   | 4 |
| 8   | 0 |
| 9   | 4 |
| 10  | 1 |
| 11  | 0 |
| 12  | 2 |

## 3. Dot Plot (Conceptual)
- Each age is marked along a horizontal axis.
- Dots stacked vertically show how many students are at each age.
- Example: Age 7 â†’ four dots stacked; Age 8 â†’ no dots.

## 4. Questions Answered with the Data
- **Most frequent age**: Ages **7 and 9**, each with 4 students.
- **Range of ages**:  
  

$$
\text{Range} = \text{Max} - \text{Min} = 12 - 5 = 7
$$


- **Number of students older than 9**:  
  Ages 10 (1 student) + 12 (2 students) = **3 students**.

## 5. Key Insight
Different representations (list, frequency table, dot plot) show the same data but make certain questions easier to answer:
- **Dot plot** â†’ quickly shows frequency and range visually.
- **Frequency table** â†’ organizes counts clearly.
- **List of numbers** â†’ raw but harder to analyze directly.


```python
# Frequency Table and Dot Plot for Student Ages

import matplotlib.pyplot as plt
from collections import Counter

# List of ages (from your example)
ages = [5, 5, 6, 7, 7, 7, 7, 9, 9, 9, 9, 10, 12, 12]

# Step 1: Construct frequency table using Counter
freq_table = Counter(ages)

# Step 2: Print frequency table
print("Frequency Table:")
print("Age | Frequency")
for age in range(min(ages), max(ages) + 1):
    print(f"{age:>3} | {freq_table.get(age, 0)}")

# Step 3: Create dot plot
plt.figure(figsize=(8, 4))
for age, count in freq_table.items():
    plt.plot([age] * count, list(range(1, count + 1)), 'o', markersize=8)

plt.xticks(range(min(ages), max(ages) + 1))
plt.xlabel("Age")
plt.ylabel("Number of Students (dots stacked)")
plt.title("Dot Plot of Student Ages")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```
