# üê¢ Stem-and-Leaf Plot: Turtle Counts at Zoos

## Problem
The government created a stem-and-leaf plot showing the number of turtles at each major zoo in the country.

### Stem-and-Leaf Plot

| Stem | Leaf           |
|------|----------------|
| 0    | 4              |
| 1    | 1 4 6 6 7 8    |
| 2    | 3 3 6 9        |
| 3    | 1 2 4 5 7 8    |
| 4    |                |
| 5    |                |
| 6    |                |

Each entry represents the number of turtles at a zoo.  
- **Stem** = tens digit  
- **Leaf** = ones digit  
- Example: Stem 2 and Leaf 3 ‚Üí 23 turtles

## Question
**How many zoos have fewer than 46 turtles?**

## Step-by-Step Solution

We interpret each row of the plot:

- **0 | 4** ‚Üí 1 zoo with 4 turtles
- **1 | 1 4 6 6 7 8** ‚Üí 6 zoos with 11, 14, 16, 16, 17, 18 turtles
- **2 | 3 3 6 9** ‚Üí 4 zoos with 23, 23, 26, 29 turtles
- **3 | 1 2 4 5 7 8** ‚Üí 6 zoos with 31, 32, 34, 35, 37, 38 turtles
- **4 |** ‚Üí No zoos listed with 40‚Äì49 turtles

Now we count all zoos with turtle counts **less than 46**:
- All values from stems 0, 1, 2, and 3 are below 46.

### Total Zoos:
- 1 (from stem 0)
- 6 (from stem 1)
- 4 (from stem 2)
- 6 (from stem 3)



$$
1 + 6 + 4 + 6 = 17
$$



## ‚úÖ Final Answer
**17 zoos** have fewer than **46 turtles**.


```python
# Turtle Count Stem-and-Leaf Plot Analysis

# Step 1: Reconstruct turtle counts from stem-and-leaf plot
turtle_counts = [
    4,                      # Stem 0
    11, 14, 16, 16, 17, 18, # Stem 1
    23, 23, 26, 29,         # Stem 2
    31, 32, 34, 35, 37, 38  # Stem 3
    # Stems 4, 5, 6 are empty
]

# Step 2: Count zoos with fewer than 46 turtles
count_under_46 = sum(1 for count in turtle_counts if count < 46)

# Step 3: Print results
print("Total number of zoos:", len(turtle_counts))
print("Number of zoos with fewer than 46 turtles:", count_under_46)

# Step 4: Optional - Plot histogram
import matplotlib.pyplot as plt
from collections import Counter

freq_table = Counter(turtle_counts)

plt.figure(figsize=(8, 5))
plt.bar(freq_table.keys(), freq_table.values(), color='green', edgecolor='black')

plt.xlabel("Number of Turtles")
plt.ylabel("Number of Zoos")
plt.title("Distribution of Turtle Counts at Major Zoos")
plt.xticks(sorted(freq_table.keys()))
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.show()
```

# üîç What this program does:

* Reconstructs the turtle counts from the stem-and-leaf plot.
* Counts how many zoos have fewer than 46 turtles.
* Optionally plots a histogram to visualize the distribution.

üü¢ When you run this, it will confirm that **17 zoos** have fewer than **46 turtles**, matching the stem-and-leaf plot interpretation.
