# ü•ß Summary: Cherry Pie Histogram

## 1. Problem
A cherry pie store notices variability in the number of cherries per pie.  
- Some pies have fewer than 50 cherries.  
- Others have over 100 cherries.  
**Goal:** Understand the distribution of pies by cherry count.

## 2. Buckets (Ranges of Cherries per Pie)
To simplify, pies are grouped into bins of 30 cherries:

| Cherry Range | Number of Pies |
|--------------|----------------|
| 0‚Äì29         | 0 |
| 30‚Äì59        | 5 |
| 60‚Äì89        | 8 |
| 90‚Äì119       | 4 |
| 120‚Äì149      | 3 |

*(No pies have more than 149 cherries.)*

## 3. Total Number of Pies


$$
5 + 8 + 4 + 3 = 20
$$

  
‚û°Ô∏è There are **20 pies** in the store.

## 4. Example Questions Answered
- **How many pies have 60 or more cherries?**  
  - Buckets: 60‚Äì89 (8), 90‚Äì119 (4), 120‚Äì149 (3).  
  - Total = $ 8 + 4 + 3 = 15$.  
  - Since there are 20 pies total, this means **all but 5 pies** have 60+ cherries.

- **Difference between pies with 60‚Äì89 cherries vs. 120‚Äì149 cherries?**  
  - 60‚Äì89 ‚Üí 8 pies  
  - 120‚Äì149 ‚Üí 3 pies  
  - Difference = $8 - 3 = 5$.  
  - ‚û°Ô∏è There are **5 more pies** in the 60‚Äì89 category.

## 5. Key Insight
- A **histogram** groups continuous data into ranges (bins) and shows how many items fall into each.  
- It allows you to quickly answer questions about totals, comparisons, and distributions.  
- In this case, the histogram reveals that most pies have **60‚Äì89 cherries**, while very few have **120‚Äì149 cherries**.


```python
# Cherry Pie Histogram Example
import matplotlib.pyplot as plt

# Step 1: Define the buckets (ranges of cherries per pie)
# Buckets: 0‚Äì29, 30‚Äì59, 60‚Äì89, 90‚Äì119, 120‚Äì149
buckets = ["0‚Äì29", "30‚Äì59", "60‚Äì89", "90‚Äì119", "120‚Äì149"]

# Step 2: Define the counts for each bucket (from the example)
counts = [0, 5, 8, 4, 3]

# Step 3: Print frequency table
print("Cherry Pie Frequency Table")
print("Range     | Number of Pies")
for b, c in zip(buckets, counts):
    print(f"{b:<8} | {c}")

# Step 4: Plot histogram
plt.figure(figsize=(8, 5))
plt.bar(buckets, counts, color=['magenta', 'blue', 'green', 'orange', 'red'], edgecolor='black')

# Step 5: Labeling
plt.xlabel("Cherry Count Range")
plt.ylabel("Number of Pies")
plt.title("Histogram of Cherry Pies by Cherry Count")
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Step 6: Show plot
plt.show()
```
