# ðŸŽ“ Median vs Mean Income by Major

## 1. Geology Majors (UNC)
- **Fact:** Michael Jordan was a geology major, earning tens of millions.  
- **Effect:** His extreme salary is an **outlier** that pulls the **mean upward**.  
- **Median:** Unaffected by the outlier, stays near the middle of typical salaries.  
- âœ… **Median < Mean**

---

## 2. Civil Engineering Majors (University of Montana)
- **Fact:** Only 3 graduates, all took the same job with identical salaries.  
- **Effect:** No variation, no outliers.  
- **Mean = Median = Salary (e.g., $50K).**  
- âœ… **Median = Mean**

---

## 3. Finance Majors (Wesleyan University)
- **Fact:** 32 graduates got high-paying consulting jobs, 3 were unemployed (income = 0).  
- **Effect:** The unemployed graduates are **low outliers** that pull the **mean downward**.  
- **Median:** Still reflects the middle of the high-paying group.  
- âœ… **Median > Mean**

---

## âœ… Key Takeaways
- **Outliers distort the mean but not the median.**  
- **High outliers (Michael Jordan)** â†’ Median < Mean.  
- **Identical values (Montana engineers)** â†’ Median = Mean.  
- **Low outliers (unemployed finance majors)** â†’ Median > Mean.


```python
# Program to compare mean vs median incomes by major
import statistics

# Step 1: Define datasets
# Geology majors (UNC) - typical salaries plus Michael Jordan as an extreme outlier
geology = [0, 45_000, 50_000, 55_000, 60_000, 70_000, 100_000, 10_000_000]  

# Civil Engineering majors (Montana) - all identical salaries
civil_eng = [50_000, 50_000, 50_000]

# Finance majors (Wesleyan) - most with high salaries, some unemployed
finance = [100_000]*32 + [0,0,0]

# Step 2: Function to calculate mean and median
def analyze(data, label):
    mean_val = statistics.mean(data)
    median_val = statistics.median(data)
    print(f"{label}:")
    print(f"  Mean   = ${mean_val:,.2f}")
    print(f"  Median = ${median_val:,.2f}")
    if median_val < mean_val:
        print("  âž¡ Median < Mean (high outlier pulling mean upward)\n")
    elif median_val > mean_val:
        print("  âž¡ Median > Mean (low outliers pulling mean downward)\n")
    else:
        print("  âž¡ Median = Mean (no outliers, identical values)\n")

# Step 3: Run analysis
analyze(geology, "Geology (UNC)")
analyze(civil_eng, "Civil Engineering (Montana)")
analyze(finance, "Finance (Wesleyan)")
```

# ðŸ”Ž What this program does:
1. **Geology (UNC):**

    - Michael Jordanâ€™s extreme salary skews the mean upward.
    - Result: **Median < Mean**.

2. **Civil Engineering (Montana):**

    - All salaries identical.
    - Result: **Median = Mean**.

3. **Finance (Wesleyan):**

    - 3 unemployed graduates (income = 0) pull the mean downward.
    - Result: **Median > Mean**.
