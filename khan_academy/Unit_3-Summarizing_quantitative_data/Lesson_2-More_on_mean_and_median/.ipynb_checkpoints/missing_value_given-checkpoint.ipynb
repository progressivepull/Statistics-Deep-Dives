# ðŸ‘¶ Babysitting Age Problem

## 1. Scenario
- Babysitter logged ages of **6 children**.  
- One age was lost due to ink spill.  
- Known: mean age = **4 years**.  
- Five ages recorded: **5, 2, 2, 4, 8**.  
- Missing age = ?  

---

## 2. Method
- Formula:  
  

$$
  \text{Mean} = \frac{\text{Sum of ages}}{\text{Number of children}}
  $$

  

- Rearranged:  
  

$$
  \text{Total sum} = \text{Mean} \times \text{Number of children}
  $$

  

- Total sum = $4 \times 6 = 24$.  
- Known ages sum = $5 + 2 + 2 + 4 + 8 = 21$.  
- Missing age = $24 - 21 = 3$.  

---

## 3. Result
- âœ… The missing age is **3 years old**.  

---

## âœ… Key Takeaway
- If you know the **mean** and the **number of data points**, you can recover missing values.  
- Here, the mean acted as a safeguard, allowing the babysitter to reconstruct the lost age.


```python
# Program to find missing age in babysitting log
import statistics

# Step 1: Known ages (one missing)
ages_known = [5, 2, 2, 4, 8]   # five ages recorded
mean_age = 4                   # given mean
num_children = 6               # total children

# Step 2: Calculate total sum using mean
total_sum = mean_age * num_children

# Step 3: Find missing age
missing_age = total_sum - sum(ages_known)

# Step 4: Print results
print("Known ages:", ages_known)
print("Mean age:", mean_age)
print("Total children:", num_children)
print("Missing age =", missing_age)

# Step 5: Verify by reconstructing full dataset
ages_complete = ages_known + [missing_age]
print("Complete ages:", ages_complete)
print("Recalculated mean =", statistics.mean(ages_complete))
```

2. Uses the formula:

3. Subtracts the sum of known ages to find the missing age.
4. Prints the missing age and verifies by recalculating the mean.
