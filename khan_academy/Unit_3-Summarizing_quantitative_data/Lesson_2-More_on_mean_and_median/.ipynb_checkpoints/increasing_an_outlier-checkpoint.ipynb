# ðŸŽ³ Bowling Scores Analysis

## 1. Scenario
- Four friends track their all-time highest bowling scores.  
- Three scores are between **180â€“220**.  
- Adamâ€™s score was **250**, then increased to **290** after a new game.  

---

## 2. Median Analysis
- With **4 numbers**, the median is the **average of the middle two scores**.  
- Middle scores are unaffected by Adamâ€™s change (still between â€œ?â€ and 220).  
- âœ… **Median does not change**.  

---

## 3. Mean Analysis
- Mean = (sum of all scores) Ã· 4.  
- Adamâ€™s score increased from 250 â†’ 290 (+40).  
- This raises the total sum by 40, so the mean increases.  
- âœ… **Mean increases**.  

---

## 4. Final Outcome
- **Median:** No change.  
- **Mean:** Increases.  

---

## âœ… Key Takeaway
- **Median is resistant to outliers or changes at the extremes.**  
- **Mean is sensitive to changes in values, especially large increases or decreases.**  
- In this case: **Median stays the same, Mean goes up.**


```python
# Program to analyze effect of changing one score on mean and median
import statistics

# Step 1: Define scores before and after Adam's new high score
scores_before = [180, 200, 220, 250]   # Example values (Adam = 250)
scores_after  = [180, 200, 220, 290]   # Adam's new score = 290

# Step 2: Function to calculate mean and median
def analyze(scores, label):
    mean_val = statistics.mean(scores)
    median_val = statistics.median(scores)
    print(f"{label}:")
    print(f"  Scores = {scores}")
    print(f"  Mean   = {mean_val:.2f}")
    print(f"  Median = {median_val:.2f}\n")

# Step 3: Run analysis
analyze(scores_before, "Before Adam's new score")
analyze(scores_after, "After Adam's new score")

# Step 4: Compare changes
mean_change = statistics.mean(scores_after) - statistics.mean(scores_before)
median_change = statistics.median(scores_after) - statistics.median(scores_before)

print("Comparison:")
print(f"  Mean change   = {mean_change:.2f}")
print(f"  Median change = {median_change:.2f}")

if median_change == 0 and mean_change > 0:
    print("âž¡ Median stays the same, Mean increases.")
```

# ðŸ”Ž What this program does:

1. Stores scores **before and after Adamâ€™s new high score**.
2. Calculates **mean and median** for both sets.
3. Prints results clearly.
4. Shows how much each measure changed.


---
## ðŸ’¡ Running this program will confirm:

* **Median:** stays the same (middle two scores unchanged).
* **Mean:** increases (sum is larger due to Adamâ€™s higher score).
