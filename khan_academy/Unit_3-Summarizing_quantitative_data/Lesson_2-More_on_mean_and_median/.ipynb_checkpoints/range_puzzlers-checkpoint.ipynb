# üöó Car Dealership Sales Puzzle

## 1. Given Information
- **Salespeople:** 11 total  
- **Median cars sold:** 6 (the middle value when sorted)  
- **Range:** 4 (max ‚àí min = 4)  

---

## 2. Key Reasoning
- With 11 people, the **median** is the 6th value ‚Üí one salesperson sold **exactly 6 cars**.  
- If the **minimum = 6**, then the **maximum = 10** (since 10 ‚àí 6 = 4).  
- If we try to make the **maximum > 10** (e.g., 11), then the minimum must be 7 to keep the range at 4.  
- But if the minimum were 7, the median could not be 6. Contradiction.  

---

## 3. Conclusion
- **Statement:** ‚ÄúAt least one of the salespeople sold more than 10 cars.‚Äù  
- **Verdict:** ‚ùå **False**  
  - The range and median together cap the maximum at **10**, so no one sold more than 10.  

---

## ‚úÖ Key Takeaway
- The **median** locks in one value at 6.  
- The **range** limits the spread to 4.  
- Together, these constraints prove that **no salesperson sold more than 10 cars**.


```python
# Program to analyze car dealership sales puzzle
import statistics

# Step 1: Define constraints
num_salespeople = 11
median_value = 6
range_value = 4

# Step 2: Construct a possible dataset
# We know the median must be 6, so one salesperson sold exactly 6 cars.
# The minimum and maximum must differ by 4.
# Example dataset consistent with constraints:
sales = [4, 5, 5, 5, 6, 6, 7, 7, 8, 9, 10]

# Step 3: Calculate median and range
median_calc = statistics.median(sales)
range_calc = max(sales) - min(sales)

print("Dataset:", sales)
print("Median =", median_calc)
print("Range =", range_calc)

# Step 4: Check if anyone sold more than 10
more_than_10 = any(s > 10 for s in sales)

print("Did anyone sell more than 10 cars?", more_than_10)

# Step 5: Logical conclusion
if more_than_10:
    print("‚û° Statement TRUE: At least one person sold more than 10 cars.")
else:
    print("‚û° Statement FALSE: No one sold more than 10 cars.")
```

# üîé What this program does:

* Sets up the **constraints**: 11 salespeople, median = 6, range = 4.
* Builds a sample dataset consistent with those constraints.
* Calculates the **median** and **range** to verify.
* Checks if anyone sold more than 10 cars.
* Prints the conclusion.


----
## üí° Running this program will confirm:

* Median = 6
* Range = 4
* No salesperson sold more than 10 cars
* ‚úÖ The statement ‚ÄúAt least one sold more than 10 cars‚Äù is **FALSE**.
