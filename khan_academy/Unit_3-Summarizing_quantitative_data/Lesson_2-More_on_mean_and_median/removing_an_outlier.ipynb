# â›³ Ana's Golf Score Analysis

## 1. Original Data (5 rounds)
- Scores: **80, 90, 92, 94, 96**  
- **Median:** Middle value = **92**  
- **Mean:** $(80 + 90 + 92 + 94 + 96) Ã· 5 = 452 Ã· 5 = 90.4$  

---

## 2. After Removing the 80 (4 rounds)
- Scores: **90, 92, 94, 96**  
- **Median:** Average of middle two (92 and 94) = **93**  
- **Mean:** $(90 + 92 + 94 + 96) Ã· 4 = 372 Ã· 4 = 93$  

---

## 3. Comparison
- **Median:** Increased from 92 â†’ 93 (+1)  
- **Mean:** Increased from 90.4 â†’ 93 (+2.6)  
- âœ… Both increased, but the **mean increased more than the median**.  

---

## âœ… Key Takeaway
- Removing a **low outlier** raises both the mean and median.  
- The **mean is more sensitive** to extreme values, so it increased more than the median.


```python
# Program to analyze Ana's golf scores (mean and median before/after removing lowest score)
import statistics

# Step 1: Define scores
scores_with_80 = [80, 90, 92, 94, 96]   # original 5 rounds
scores_without_80 = [90, 92, 94, 96]    # after removing the 80

# Step 2: Function to calculate mean and median
def analyze(scores, label):
    mean_val = statistics.mean(scores)
    median_val = statistics.median(scores)
    print(f"{label}:")
    print(f"  Scores = {scores}")
    print(f"  Mean   = {mean_val:.2f}")
    print(f"  Median = {median_val:.2f}\n")

# Step 3: Run analysis
analyze(scores_with_80, "With 80 included")
analyze(scores_without_80, "Without 80 (removed)")

# Step 4: Compare changes
mean_change = statistics.mean(scores_without_80) - statistics.mean(scores_with_80)
median_change = statistics.median(scores_without_80) - statistics.median(scores_with_80)

print("Comparison:")
print(f"  Mean increased by {mean_change:.2f}")
print(f"  Median increased by {median_change:.2f}")
print("âž¡ Both increased, but the mean increased more than the median.")
```
