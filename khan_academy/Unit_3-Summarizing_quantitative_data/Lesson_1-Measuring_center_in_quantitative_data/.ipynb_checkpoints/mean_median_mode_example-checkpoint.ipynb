# üìä Mean, Median, and Mode Calculation

## Dataset
Numbers:  
`23, 29, 20, 32, 23, 21, 33, 25`

---

## 1. Mean (Arithmetic Average)
- Formula:  
  

$$
\text{Mean} = \frac{\text{Sum of all values}}{\text{Number of values}}
$$


- Sum = $23 + 29 + 20 + 32 + 23 + 21 + 33 + 25 = 206$  
- Count = 8  
- Mean = $206 / 8 = 25.75$  

‚û°Ô∏è **Mean = 25.75**

---

## 2. Median (Middle Value)
- Step 1: Order the data ‚Üí  
  $20, 21, 23, 23, 25, 29, 32, 33$  
- Step 2: Since there are 8 values (even count), take the average of the two middle numbers.  
- Middle values = 23 and 25  
- Median = $(23 + 25) / 2 = 24$  

‚û°Ô∏è **Median = 24**

---

## 3. Mode (Most Frequent Value)
- Mode = the number that appears most often.  
- In this dataset, **23** appears twice.  
- All other numbers appear once.  

‚û°Ô∏è **Mode = 23**

---

## ‚úÖ Final Results
- **Mean:** 25.75  
- **Median:** 24  
- **Mode:** 23


```python
# Program to calculate Mean, Median, and Mode
import statistics

# Step 1: Define the dataset
data = [23, 29, 20, 32, 23, 21, 33, 25]

# Step 2: Calculate Mean
mean_value = statistics.mean(data)

# Step 3: Calculate Median
median_value = statistics.median(data)

# Step 4: Calculate Mode
try:
    mode_value = statistics.mode(data)
except statistics.StatisticsError:
    mode_value = "No unique mode"

# Step 5: Print results
print("Dataset:", data)
print("Mean (Arithmetic Average):", mean_value)
print("Median (Middle Value):", median_value)
print("Mode (Most Frequent Value):", mode_value)
```
