# üìä Introduction to Statistics

## 1. What is Statistics?
Statistics is the study of data ‚Äî collecting, organizing, analyzing, and interpreting it.  
Two main branches:
- **Descriptive statistics**: Summarizing data with representative numbers or visuals.  
- **Inferential statistics**: Making predictions or conclusions based on data.

---

## 2. Descriptive Statistics
- Goal: Represent a dataset with a **smaller set of numbers** that capture its essence.  
- Example: Heights of plants ‚Üí instead of listing all values, describe them with one ‚Äútypical‚Äù number.

---

## 3. Measures of Central Tendency
These are ways to find a ‚Äútypical‚Äù or ‚Äúmiddle‚Äù value in a dataset.

### üîπ Arithmetic Mean (Average)
- Definition:  
  

\[
  \text{Mean} = \frac{\text{Sum of all values}}{\text{Number of values}}
  \]


- Example dataset: $4, 3, 1, 6, 1, 7$  
  - Sum = 22  
  - Count = 6  
  - Mean = $22 / 6 = 3.67$ (‚âà 3 and 2/3)  
- Represents the **balance point** of the data.

---

### üîπ Median
- Definition: The **middle value** when data is ordered.  
- If even number of values ‚Üí average of the two middle numbers.  
- Example dataset ordered: $1, 1, 3, 4, 6, 7$  
  - Middle two = 3 and 4  
  - Median = $3.5$  
- Useful when data has **extreme outliers** (less affected than the mean).

---

### üîπ Mode
- Definition: The **most frequent value** in the dataset.  
- Example dataset: $4, 3, 1, 6, 1, 7$  
  - Mode = 1 (appears twice).  
- Useful when one value occurs much more often than others.

---

## 4. Key Insights
- **Mean**: Best for balanced data without extreme outliers.  
- **Median**: Best when data is skewed or has extreme values.  
- **Mode**: Best when identifying the most common or popular value.  
- Each measure gives a different perspective on what is ‚Äútypical.‚Äù

---

## ‚úÖ Summary
Descriptive statistics help us simplify complex datasets into meaningful numbers.  
Measures of central tendency ‚Äî mean, median, and mode ‚Äî are foundational tools for understanding data distributions and preparing for deeper statistical analysis.


```python
# Measures of Central Tendency Example
import statistics

# Step 1: Dataset (plant heights in inches)
data = [4, 3, 1, 6, 1, 7]

# Step 2: Calculate Mean
mean_value = statistics.mean(data)

# Step 3: Calculate Median
median_value = statistics.median(data)

# Step 4: Calculate Mode
try:
    mode_value = statistics.mode(data)
except statistics.StatisticsError:
    mode_value = "No unique mode"

# Step 5: Print results
print("Dataset:", data)
print("Mean (Arithmetic Average):", mean_value)
print("Median (Middle Value):", median_value)
print("Mode (Most Frequent Value):", mode_value)
```
