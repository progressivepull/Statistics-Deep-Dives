# üìä Exam Scores Analysis

## Final Exam (Yellow Bars)
- **Scores:**  
  - Ishaan: 100  
  - Emily: 100  
  - Daniel: 100  
  - Jessica: 75  
  - William: 80  

- **Median:** 100  
  (Ordered scores: 75, 80, 100, 100, 100 ‚Üí middle value is 100)

- **Average (Mean):**  
  
$$
\frac{100 + 100 + 100 + 75 + 80}{5} = \frac{455}{5} = 91
$$

- **Mode:** 100  
  (Appears 3 times, most frequent score)

---

## Midterm Exam (Blue Bars)
- **Scores:**  
  - Jessica: 100  
  - Daniel: 60  
  - (others not explicitly listed in transcript)

- **Midrange:**  
  
$$
\frac{100 + 60}{2} = 80
$$

- **Range:**  
  
$$
100 - 60 = 40
$$

---

## üìù Key Takeaways
- The **final exam** was relatively easy, with most students scoring 100.  
- The **median, mode, and average** all highlight strong performance on the final.  
- The **midterm** had more spread, with scores ranging from 60 to 100.  
- **Midrange (80)** and **range (40)** show greater variability compared to the final exam.
