# Gender and Voting Preferences in the 2012 U.S. Presidential Election

## Two-Way Table of Column Relative Frequencies
| Gender | Obama | Romney | Other | Total |
|--------|-------|--------|-------|-------|
| Men    | 42%   | 52%    | 6%    | 100%  |
| Women  | 52%   | 43%    | 5%    | 100%  |

## Analysis
- **Men:** 52% voted for Romney.  
- **Women:** 43% voted for Romney.  
- Since **52% > 43%**, male voters were more likely to support Romney compared to female voters.

## Conclusion
✅ **Yes — male voters were more likely to vote for Romney than female voters in the 2012 election.**
