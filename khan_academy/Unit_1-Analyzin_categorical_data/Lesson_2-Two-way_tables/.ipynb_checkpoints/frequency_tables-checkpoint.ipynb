---

## âœ¨ Why Relative Frequencies Matter

- Relative frequencies allow comparison across categories by normalizing totals to **1.00**.  
- Even though SUVs and sports cars had different sample sizes, percentages show that **sports car drivers had a slightly higher accident rate (25% vs. 22%)**.  
- This perspective highlights proportions rather than raw counts, making comparisons clearer.
