# Computer Use vs Sleep Study Summary

## Step 1: Total Participants
- Moderate computer users with 5–7 hours of sleep = **17 people**.
- This group represents **10% of the total**.
- Total participants = 17 ÷ 0.10 = **170**.

## Step 2: Extreme Users, 5–7 Hours
- Extreme computer users with 5–7 hours = **11.7% of total**.
- Calculation: 0.117 × 170 = 19.89 ≈ **20 people**.

**Answer:** 20 people were extreme computer users who slept 5–7 hours.

---

## Step 3: Association Between Minimal Use and ≥7 Hours Sleep
- Minimal computer users:
  - **51% sleep ≥7 hours**.
  - Only **16% sleep ≤5 hours**.
- Extreme computer users:
  - **47% sleep ≤5 hours**.
  - Only **19% sleep ≥7 hours**.
- Moderate users fall in between.

---

## Evidence of Association
- **51% of minimal users sleep ≥7 hours vs 33% overall.**
- **55% of those who sleep ≥7 hours are minimal users, though only 35% overall are minimal users.**

**Conclusion:** Yes, there is evidence of association. Minimal computer use is linked to more sleep.

---

## Final Selected Explanations
- ✔ Yes, because **51% of minimal users sleep ≥7 hours compared to 33% overall.**
- ✔ Yes, because **55% of ≥7-hour sleepers are minimal users, though only 35% overall are minimal users.**


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.simplefilter(action="ignore", category=FutureWarning)

# Load the Excel file
df = pd.read_excel("computer_sleep_study.xlsx")

# Bin sleep hours into categories
def categorize_sleep(hours):
    if hours <= 5:
        return "≤5"
    elif 5 < hours <= 7:
        return "5–7"
    else:
        return "≥7"

df["SleepCategory"] = df["SleepHours"].apply(categorize_sleep)

# Create contingency table
contingency = pd.crosstab(df["ComputerUse"], df["SleepCategory"])

# Calculate percentages
total = contingency.to_numpy().sum()
rowPercent = contingency.div(contingency.sum(axis=1), axis=0) * 100
columnPercent = contingency.div(contingency.sum(axis=0), axis=1) * 100
totalPercent = contingency / total * 100

# Create annotated labels
annot = contingency.copy()
for row in contingency.index:
    for col in contingency.columns:
        r = rowPercent.loc[row, col]
        c = columnPercent.loc[row, col]
        t = totalPercent.loc[row, col]
        annot.loc[row, col] = f"{contingency.loc[row, col]}\nR:{r:.1f}%\nC:{c:.1f}%\nT:{t:.1f}%"

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(contingency, annot=annot.values, fmt="", cmap="YlGnBu", cbar=False,
            xticklabels=contingency.columns, yticklabels=contingency.index)
plt.title("Computer Use vs Sleep Hours\n(Row %, Column %, Total %)")
plt.xlabel("Sleep Hours")
plt.ylabel("Computer Use")
plt.tight_layout()
plt.savefig("Computer_time.png")
plt.show()
```
