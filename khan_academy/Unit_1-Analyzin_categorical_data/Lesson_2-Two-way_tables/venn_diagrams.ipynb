# Chocolate and Coconut Classification

We have **12 chocolates** in total.  
- **Brown color** = chocolate coating on the outside  
- **Letter C** = coconut filling inside  

Examples:
- Chocolate outside, no coconut inside  
- Chocolate outside **and** coconut inside  
- No chocolate outside, but coconut inside  
- Neither chocolate nor coconut  

---

## üç´ Venn Diagram Representation

- **Universe (rectangle)** = all 12 chocolates  
- **Circle 1** = chocolates with chocolate coating  
- **Circle 2** = chocolates with coconut filling  

### Counts
- Chocolate but no coconut = **6**  
- Chocolate and coconut = **3**  
- Coconut but no chocolate = **1**  
- Neither chocolate nor coconut = **2**

### Totals
- Chocolates = 6 + 3 = **9**  
- Coconuts = 1 + 3 = **4**  
- Neither = **2**

---

## üìã Two-Way Table Representation

|                | Coconut | No Coconut | Total |
|----------------|---------|------------|-------|
| **Chocolate**  | 3       | 6          | 9     |
| **No Chocolate** | 1     | 2          | 3     |
| **Total**      | 4       | 8          | 12    |

### Breakdown
- Coconut total = 3 + 1 = **4**  
- No coconut total = 6 + 2 = **8**  
- Chocolate total = 3 + 6 = **9**  
- No chocolate total = 1 + 2 = **3**

---

## ‚úÖ Key Insight
Both the **Venn diagram** and the **two-way table** represent the same distribution of chocolates, just in different visual formats.
