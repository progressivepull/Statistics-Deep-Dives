# üèÄ Summary: Basketball Team Stem-and-Leaf Plot

## 1. Problem
A statistician tracked the **points scored by 12 players** in one game.  
To visualize the distribution, they created a **stem-and-leaf plot** (also called a stem plot).  
**Question:** How many total points did the team score?

## 2. How to Read a Stem-and-Leaf Plot
- **Stem** = tens digit (e.g., ‚Äú1‚Äù means scores in the teens).  
- **Leaf** = ones digit (e.g., ‚Äú3‚Äù means the score is 13).  
- Together, stem + leaf = full score.  
- This format shows both **individual scores** and the **distribution** across ranges.

## 3. Extracted Player Scores
From the stem-and-leaf plot, the scores were:

- **0‚Äì9 range (stem = 0):**  
  0, 2, 4, 7, 7, 9, 9  
- **10‚Äì19 range (stem = 1):**  
  11, 11, 13, 18  
- **20‚Äì29 range (stem = 2):**  
  20  

‚û°Ô∏è Total of **12 scores** (matching 12 players).

## 4. Distribution
- **7 players** scored between 0‚Äì9 points.  
- **4 players** scored between 10‚Äì19 points.  
- **1 player** scored 20 points.  

This shows most players scored in the **single digits**, fewer in the teens, and only one reached 20.

## 5. Total Team Points
Adding all scores:  

$$
0 + 2 + 4 + 7 + 7 + 9 + 9 + 11 + 11 + 13 + 18 + 20 = 102
$$


‚û°Ô∏è The team scored a total of **102 points**.

## 6. Key Insight
- A **stem-and-leaf plot** is useful because it shows both the **raw data values** and the **distribution** at the same time.  
- It‚Äôs easy to see how many players scored in each range and to calculate totals.  
- In this case, the visualization highlights that most players contributed modestly, with only one standout at 20 points.


```python
# Basketball Team Stem-and-Leaf Plot Example
# Calculate total points scored by the team

# Step 1: Recreate the scores from the stem-and-leaf plot
scores = [0, 2, 4, 7, 7, 9, 9, 11, 11, 13, 18, 20]

# Step 2: Print frequency table
from collections import Counter
freq_table = Counter(scores)

print("Frequency Table:")
print("Score | Frequency")
for score in sorted(freq_table.keys()):
    print(f"{score:>5} | {freq_table[score]}")

# Step 3: Calculate total points
total_points = sum(scores)
print("\nTotal Team Points:", total_points)

# Step 4: Optional - Plot histogram of scores
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.bar(freq_table.keys(), freq_table.values(), color='skyblue', edgecolor='black')

plt.xlabel("Player Score")
plt.ylabel("Number of Players")
plt.title("Basketball Team Scores (Stem-and-Leaf Data)")
plt.xticks(sorted(freq_table.keys()))
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.show()
```

# üåø Stem-and-Leaf Plots: Benefits and Uses

## 1. Benefits
- **Preserve Original Data**: Unlike histograms, stem-and-leaf plots display the actual data values, not just summarized counts.
- **Show Distribution Clearly**: They reveal the shape of the dataset (symmetry, skewness, spread) at a glance.
- **Spot Outliers Quickly**: Extreme values stand out easily, helping detect unusual observations.
- **Sorted Representation**: Leaves are typically arranged in ascending order, so the dataset is automatically sorted, aiding in finding medians or quartiles.
- **Easy to Construct by Hand**: Practical for classroom settings or quick exploratory analysis without software.
- **Effective for Small/Medium Data**: Best suited for datasets with fewer than ~50‚Äì100 observations; larger datasets may look cluttered.

## 2. Common Uses
| Use Case | How Stem-and-Leaf Helps |
|----------|--------------------------|
| **Exploratory Data Analysis** | Provides a quick view of central tendency, variability, and distribution shape |
| **Quality Control** | Used to assess process variation, detect defects, and summarize inspection data |
| **Comparing Groups** | Side-by-side plots can compare distributions across different samples or batches |
| **Teaching Statistics** | Helps students understand how raw data connects to frequency distributions and histograms |
| **Identifying Outliers** | Makes unusual values stand out visually |

## 3. Limitations
- Not ideal for very large datasets (over 100 values) ‚Äî histograms or box plots are clearer in those cases.
- Less common in modern software, though still valuable for teaching and exploratory analysis.

## ‚úÖ Summary
Stem-and-leaf plots are a simple but powerful tool for small datasets.  
They combine the clarity of a histogram with the detail of raw data, making them excellent for spotting distribution patterns, outliers, and teaching statistical concepts.
