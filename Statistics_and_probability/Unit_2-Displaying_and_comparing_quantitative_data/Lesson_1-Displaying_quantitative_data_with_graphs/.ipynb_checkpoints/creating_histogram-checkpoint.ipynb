# üçΩÔ∏è Summary: Visualizing Ages in a Restaurant with Histograms

## 1. Problem
You collect the ages of everyone in a restaurant.  
Looking at raw numbers doesn‚Äôt reveal much about the distribution (young vs. middle-aged vs. seniors).  
**Goal:** Group ages into categories (bins) and visualize them.

## 2. Buckets (10-Year Ranges)
| Age Range | Number of People |
|-----------|------------------|
| 0‚Äì9       | 6 |
| 10‚Äì19     | 3 |
| 20‚Äì29     | 5 |
| 30‚Äì39     | 1 |
| 40‚Äì49     | 2 |
| 50‚Äì59     | 2 |
| 60‚Äì69     | 1 |

## 3. Histogram
- **X-axis** ‚Üí Age buckets (0‚Äì9, 10‚Äì19, ‚Ä¶, 60‚Äì69)  
- **Y-axis** ‚Üí Number of people in each bucket  
- Each bar‚Äôs height shows how many people fall into that age range.

## 4. Insights
- The **largest group** is ages **0‚Äì9** (6 people).  
- Ages **20‚Äì29** also have a strong presence (5 people).  
- Very few people in the **30‚Äì39** and **60‚Äì69** ranges (only 1 each).  
- Overall, the restaurant skews **young and family-friendly**, with many children and fewer seniors.

## 5. Key Idea
A **histogram** is useful when:
- Data can take many values.
- You want to group values into ranges (bins).
- You want to see the **distribution** across categories.

Unlike a dot plot (which shows each individual), a histogram summarizes groups, making patterns clearer.


```python
# Histogram of Ages in a Restaurant
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Raw ages (example dataset from the scenario)
ages = [2, 4, 5, 7, 8, 9, 12, 15, 18, 22, 25, 27, 29,
        33, 42, 45, 52, 55, 63]  # You can adjust this list

# Step 2: Define bins (10-year ranges)
bins = [0, 10, 20, 30, 40, 50, 60, 70]

# Step 3: Plot histogram
plt.figure(figsize=(8, 5))
plt.hist(ages, bins=bins, edgecolor='black', color='skyblue')

# Step 4: Labeling
plt.xticks(bins)
plt.xlabel("Age Range")
plt.ylabel("Number of People")
plt.title("Histogram of Ages in Restaurant")
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Step 5: Show plot
plt.show()
```
